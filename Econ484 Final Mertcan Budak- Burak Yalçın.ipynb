```python
import numpy as np
import pandas as pd 
```

```python
MC=pd.read_csv("marketing_campaign.csv",sep="\t")
```

```python
MC.head()
```

```python
MC.describe().T
```

```python
MC['Age'] = 2022 - MC['Year_Birth']
MC.head()
```

```python
MC.drop(['Z_CostContact' ,'Z_Revenue','AcceptedCmp3','AcceptedCmp1','AcceptedCmp2','AcceptedCmp4','AcceptedCmp5','Complain','Response'] , axis = 1 , inplace=True)
```

```python
MC.head()
```

```python
MC['Marital_Status'] = MC['Marital_Status'].replace(['Married', 'Together'],'Relationship')
MC['Marital_Status'] = MC['Marital_Status'].replace(['Divorced', 'Widow', 'Alone', 'YOLO', 'Absurd'],'Single')
```

```python
MC['Age'].value_counts()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.NaN,
                        strategy ='mean')
                    
imputer = imputer.fit(MC[['Income']])
MC['Income'] = imputer.transform(MC[['Income']])
MC
```

```python
MC["Total_spend"] = MC["MntWines"]+ MC["MntFruits"]+ MC["MntMeatProducts"]+ MC["MntFishProducts"]+ MC["MntSweetProducts"]+ MC["MntGoldProds"]
```

```python
MC = MC.rename(columns={"MntWines": "Wines","MntFruits":"Fruits","MntMeatProducts":"Meat","MntFishProducts":"Fish","MntSweetProducts":"Sweets","MntGoldProds":"Gold"})
MC
```

```python
MC.drop(['ID','Year_Birth','Dt_Customer','Recency'] , axis = 1 , inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=MC, x='Fish')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder =LabelEncoder()

MC['Marital_Status']= label_encoder.fit_transform(MC['Marital_Status'])
MC.head(10)
```

```python
mc_copy = MC.copy()
cat_col =  [x for x in mc_copy.columns if ((mc_copy[x].dtype != 'int64') & (mc_copy[x].dtype != 'float64'))]
mc_copy=pd.get_dummies(mc_copy , columns=cat_col , drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler
num_cols =  [x for x in mc_copy.columns if ~((mc_copy[x].dtype != 'int64') & (mc_copy[x].dtype != 'float64'))]
Scaler = StandardScaler()
mc_copy[num_cols] = Scaler.fit_transform(mc_copy[num_cols])
```

```python
    from sklearn.cluster import KMeans
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 0.95)
pca.fit(mc_copy)
reduced_df= pd.DataFrame(pca.transform(mc_copy))

print("Current number of features= ",len(reduced_df.columns))
```

```python
pca = PCA(n_components = 0.75)
pca.fit(mc_copy)
reduced_df= pd.DataFrame(pca.transform(mc_copy))

print("Current number of features= ",len(reduced_df.columns))
```

```python
distortions = []
K = range(1,7)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(reduced_df)
    distortions.append(kmeanModel.inertia_)
```

```python
sns.set_palette(sns.color_palette("RdPu", 6))
palette = ["#fcd7d3","#faabb8","#f667a0","#cd238e","#cd238e"]
```

```python
plt.figure(figsize=(15,5))
plt.rcParams.update({'axes.grid': True})
plt.plot(K, distortions, 'bx-', color=palette[-1])
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method Showing The Optimal k')
sns.despine()
plt.show()
```

```python
clusters = KMeans(n_clusters=4, random_state=202).fit(reduced_df)
MC["Personality"]= clusters.labels_
mc_copy["Personality"]= clusters.labels_
```

```python
plt.figure(figsize = (16,8) )
sns.kdeplot(data =MC, x = 'Marital_Status' , hue = 'Personality' , color=palette[-1])
plt.title('Marital Status' ,fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (14,8) )
sns.kdeplot(data =MC, x = 'Fish' , hue = 'Personality' ,color=palette[-1])
plt.title('Fish Distribution' ,fontsize = 15)
plt.show()
```
